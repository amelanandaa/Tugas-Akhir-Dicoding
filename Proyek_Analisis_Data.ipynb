### Pertanyaan 3: Bagaimana performa pendapatan setiap bulannya?

```python
monthly = all_df.resample(rule='M', on="order_purchase_timestamp").agg({
    "order_id":"nunique",
    "payment_value":"sum"
})
monthly.index = monthly.index.strftime('%Y-%m')
monthly.head()
```

```python
monthly = all_df.resample(rule='M', on="order_purchase_timestamp").agg({
    "order_id":"nunique",
    "payment_value":"sum"
})
monthly.index = monthly.index.strftime('%Y-%m')

plt.figure(figsize=(20, 5))
plt.plot( monthly.index, monthly["payment_value"], color="#72BCD4")
plt.title("Total Revenue per Month", loc="center", fontsize=20)
plt.xlabel('Date', fontsize=10)
plt.ylabel('Revenue',fontsize=10)
plt.xticks(rotation=45)
plt.show()
```

### RFM Analysis

```python
rfm_df = all_df.groupby(by="customer_id", as_index=False).agg({
    "order_purchase_timestamp" : "max",
    "order_id":"nunique",
    "payment_value":"sum"
})
rfm_df.columns = ["customer_id","max_order","frequency","monatary"]

rfm_df["max_order"]= rfm_df["max_order"].dt.date
recent = all_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order"].apply(lambda x: (recent - x).days)

rfm_df.drop("max_order", axis=1, inplace =True)
rfm_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30,5))
colors = ["#9BCF53","#9BCF53","#9BCF53","#9BCF53","#9BCF53"]

sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=10, rotation=45)

sns.barplot(y="frequency", x="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=10, rotation=45)

sns.barplot(y="monatary", x="customer_id", data=rfm_df.sort_values(by="monatary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=10, rotation=45)

plt.suptitle("RFM Analysis", fontsize=15)
plt.show()
```
